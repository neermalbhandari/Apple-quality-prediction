```python
#importing necessary libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#reading csv file from this link
df = pd.read_csv("/kaggle/input/apple-quality/apple_quality.csv")
```

```python
#exploring the top rows of the data
df.head()
```

```python
#exploring number of null items in the columns
df.isnull().sum()
```

```python
#exploring basic descripion of the data
df.info()
```

```python
#converting Acidity column into numeric form
df['Acidity'] = df['Acidity'].apply(pd.to_numeric, errors='coerce')
```

```python
#filling missing data
col=df.columns
for i in col:
    if df[i].dtype=='float64':
        df[i]=df[i].fillna(df[i].mean())
    else:
        df[i]=df[i].fillna(df[i].mode()[0])  
```

```python
#rechecking null values
df.isnull().sum()
```

```python
#plotting bar graph (Quality counts)
plt.figure(figsize=(6, 6))
bar = plt.bar(x=df['Quality'].value_counts().index, height=df['Quality'].value_counts().values)
plt.bar_label(bar)
plt.show()
```

```python
#Counting values in Quality column
df['Quality'].value_counts()
```

```python
#assigning numeric values to the elements of Quality column
df['Quality']=df['Quality'].map({
    'good':1,
    'bad':0
})
```

```python
#Splitting independent and dependent  variables
y=df['Quality']
yid=df['A_id']
x=df.drop(['A_id','Quality'],axis=1)
```

```python
#cheking description of dependents
x.info()
```

```python
#Checking missing values of dependents
x.isnull().sum()
```

```python
#plotting a boxplot for x
x.boxplot()
```

```python
x.isnull().sum()
```

```python
for i in x.columns:
    plt.hist(x[i], bins=100, color='pink')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#importing essential libraries for splitting the dataset into train and test sets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x.isnull().sum()
```

```python
#importing, testing models and predicting
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
y_predr = rf.predict(x_test)
import xgboost as xgbt
xgb = xgbt.XGBClassifier(random_state=42)
xgb.fit(x_train, y_train)
y_predx = xgb.predict(x_test)
import lightgbm as lgbm
lgb = lgbm.LGBMClassifier(n_estimators = 50,learning_rate = 0.1,num_leaves = 4,class_weight = 'balanced')
lgb.fit(x_train, y_train)
y_predl = lgb.predict(x_test)
from catboost import CatBoostClassifier
ctb = CatBoostClassifier(n_estimators=500, max_depth=7,boosting_type = 'Ordered', verbose=0)
ctb.fit(x_train, y_train)
y_predc = ctb.predict(x_test)
```

```python
#checking accuracy scores
accuracy_score(y_test,y_predr)
```

```python
accuracy_score(y_test,y_predx)
```

```python
accuracy_score(y_test,y_predl)
```
